```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
emails = [
    "Buy now with an exclusive discount",      # spam
    "Big lottery – only today!",              # spam
    "Meeting tomorrow at 9 AM",               # normal
    "Your purchase invoice is ready"          # normal
]

labels = [1, 1, 0, 0]  # 1=spam, 0=normal
```

```python
vec = CountVectorizer()
X = vec.fit_transform(emails)
X.shape, vec.get_feature_names_out(),vec.vocabulary_
```

```python
model = LogisticRegression()
model.fit(X, labels)
```

```python
test = ["Amazing discount today!"]
print("Spam probability:", model.predict_proba(vec.transform(test))[0][1])
```
